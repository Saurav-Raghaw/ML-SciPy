```python
#Matplotlib Pyplot

#Matplotlib utilities lies under the pyplot submodule

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Draw a line in a diagram from position (0,0) to position (6,250):

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Plotting x and y points

#The plot() function is used to draw points (markers) in a diagram.

#The function takes parameters for specifying points in the diagram.

#Parameter 1 is an array containing the points on the x-axis.

#Parameter 2 is an array containing the points on the y-axis.

#Drawing a line in a diagram from position (1, 3) to position (8, 10).

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#Plotting Without Line.
#Drawing two points in the diagram, one at position (1, 3) and one in position (8, 10).


xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10).

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, 'o')
plt.plot(xpoints, ypoints)

plt.show()
```

```python
#Default X-Points.
#If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, (etc. depending on the length of the y-points).

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o')
plt.plot(ypoints)

plt.show()
```

```python
#Markers.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
#Marker reference
'''
Marker	Description

'o'	    Circle	
'*'	    Star	
'.'	    Point	
','	    Pixel	
'x'	    X	
'X'	    X (filled)	
'+'	    Plus	
'P'	    Plus (filled)	
's'	    Square	
'D'	    Diamond	
'd'	    Diamond (thin)	
'p'	    Pentagon	
'H'	    Hexagon	
'h'	    Hexagon	
'v'	    Triangle Down	
'^'	    Triangle Up	
'<'	    Triangle Left	
'>'	    Triangle Right	
'1'	    Tri Down	
'2'	    Tri Up	
'3'	    Tri Left	
'4'	    Tri Right	
'|'    	Vline	
'_'	    Hline

'''
```

```python
#Format Strings fmt.

#We can use also use the shortcut string notation parameter to specify the marker.


#marker|line|color

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
#Line Reference
'''
Syntax	Description

'-'	    Solid line	
':'	    Dotted line	
'--'	  Dashed line	
'-.'	  Dashed/dotted line

'''
```

```python
#Color Reference
Syntax	Description
'''

'r'	   Red	
'g'	   Green	
'b'	   Blue	
'c'	   Cyan	
'm'	   Magenta	
'y'	   Yellow	
'k'	   Black	
'w'	   White

'''
```

```python
#Marker Size
#ms to set the size of the markers.

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

```python
#Marker Color.

#mec to set the color of the edge of the marker.
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
#markerfacecolor or the shorter mfc to set the color inside the edge of the markers.

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```
