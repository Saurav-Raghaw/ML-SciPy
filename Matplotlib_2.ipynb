```python
#Linestyle
#linestyle, or ls, to change the style of the plotted line.

import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
'''
Style    	Or
'solid' (default)	'-'	
'dotted'	':'	
'dashed'	'--'	
'dashdot'	'-.'	
'None'	'' or ' '

'''
```

```python
#Line Color
#color or the shorter c to set the color of the line.

plt.plot(ypoints, color = 'r')
plt.show()
```

```python
#Line Width.
#linewidth or the shorter lw to change the width of the line.

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
#Multiple Lines.

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])


plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
#Create Labels for a Plot

#xlabel() and ylabel() functions to set a label for the x- and y-axis.

plt.plot(y1)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#Create a Title for a Plot

plt.plot(y1)


plt.title("Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#we can use fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
#Position the Title
#We can use the loc parameter in title() to position the title.

#Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

```python
#Add Grid Lines to a Plot
#the grid() function to add grid lines to the plot.

plt.plot(y1)


plt.title("Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid()
plt.show()
```

```python
#We can use the axis parameter in the grid() function to specify which grid lines to display.

#Legal values are: 'x', 'y', and 'both'. Default value is 'both'.

plt.plot(y1)


plt.title("Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(axis= 'x')
plt.show()
```

```python
#We can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).


plt.plot(y1)


plt.title("Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
#subplots() function we can draw multiple plots in one figure.

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.

plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.

plt.plot(x,y)

plt.show()
```

```python
#For 2 ros and 1 columns.

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
#Title

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

```python
#Super Title
#We can add a title to the entire figure with the suptitle() function.

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
